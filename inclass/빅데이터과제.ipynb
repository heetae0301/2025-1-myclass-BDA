```python
import pandas as pd
import matplotlib.pyplot as plt

# CSV 파일 불러오기
df = pd.read_csv("서울시_민간소비_소득_통합.csv", encoding='utf-8-sig')

# x축 설정
x = df['연도']
x_index = range(len(x))
bar_width = 0.6

# 그래프 설정
fig, axes = plt.subplots(2, 1, figsize=(14, 10))  # 2행 1열 subplot

# 그래프 1: 민간소비지출액
axes[0].bar(x_index, df['민간소비지출액(천원)'], width=bar_width, color='royalblue')
axes[0].set_title('서울시 연도별 민간소비지출액')
axes[0].set_ylabel('금액 (천 원)', rotation=0, labelpad=60)
axes[0].set_xticks(x_index)
axes[0].set_xticklabels(x, rotation=45)
axes[0].grid(axis='y')

# 그래프 2: 1인당 개인소득
axes[1].bar(x_index, df['1인당_개인소득(천원)'], width=bar_width, color='orange')
axes[1].set_title('서울시 연도별 1인당 개인소득')
axes[1].set_xlabel('연도')
axes[1].set_ylabel('금액 (천 원)', rotation=0, labelpad=60)
axes[1].set_xticks(x_index)
axes[1].set_xticklabels(x, rotation=45)
axes[1].grid(axis='y')

# 전체 그래프 레이아웃 정리
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# CSV 파일 불러오기 (코랩에서는 직접 업로드 필요)
df = pd.read_csv('/content/주요지표_연간지표__20250506182224.csv', encoding='cp949')

# 연도별 데이터를 긴 형태로 변환
df_melted = df.melt(id_vars='계정항목별', var_name='년도', value_name='값')

# 특정 항목 선택 (예: 국내총생산(명목, 원화표시))
target = '국내총생산(명목, 원화표시) (십억원)'
plot_df = df_melted[df_melted['계정항목별'] == target].copy()
plot_df['년도'] = plot_df['년도'].astype(int)

# 그래프 출력
plt.figure(figsize=(12, 6))
plt.plot(plot_df['년도'], plot_df['값'], marker='o', linewidth=2)
plt.title(f'{target} 변화 추이', fontsize=14)
plt.xlabel('년도')
plt.ylabel('값 (십억원)')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# CSV 파일 불러오기
df = pd.read_csv('/content/주요지표_연간지표__20250506182224.csv', encoding='cp949')

# 데이터 정리 (긴 형식으로 변환)
df_melted = df.melt(id_vars='계정항목별', var_name='년도', value_name='값')
df_melted['년도'] = df_melted['년도'].astype(int)

# 각각 따로 그래프 그릴 항목
targets = {
    '국내총생산(명목, 원화표시) (십억원)': '국내총생산 ',
    '국민총소득(명목, 원화표시) (십억원)': '국민총소득 '
}

# 각각 그래프 생성
for indicator, title in targets.items():
    data = df_melted[df_melted['계정항목별'] == indicator]
    plt.figure(figsize=(12, 6))
    plt.plot(data['년도'], data['값'], marker='o', color='steelblue')
    plt.title(title, fontsize=14)
    plt.xlabel('년도')
    plt.ylabel('값 (십억원)', rotation=0, labelpad=50)  # y축 제목 가로로 설정
    plt.grid(True)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```
