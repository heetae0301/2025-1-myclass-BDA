```python
foreigner['code']=foreigner['code'].astype(str)
foreigner.info()
```

```python
bins=[0,1000,3000,5000,10000,15000,20000, 25000,30000]
```

```python
from folium import Map
map_seoul=Map(location=[37,55,127],
              zoom_start=12,
              tiles='cartodbpositron')
map_seoul
```

```python
from folium import Choropleth

cp = Choropleth(
    geo_data = geo_seoul,
    data = foreigner,
    columns = ('code', 'pop'),
    key_on = 'feature.properties.AMD_DR_CD',
    fill_color = 'Blues',
    nan_fill_color = 'White',
    fill_opacity = 1,
    line_opacity = 0.5,
    bins = bins)

cp.add_to(map_seoul)
map_seoul
```

```python
from IPython.display import display_html
def display_side_by_side(*args):
    """여러 데이터프레임 비교가 쉽게 옆쪽으로 표시한다"""
    html_str=''
    for df in args:
        html_str += df.to_html() + '&nbsp;'*4
    display_html(html_str.replace('table','table style="display:inline"'), raw=True)
```

```python
df
```

```python
df.apply(lambda x: x+5)
```

```python
display_side_by_side(df, df.apply(lambda x: x*5))
```

```python
display_side_by_side(df, df.apply(lambda x: x*5,axis=1))
```

```python
f=lambda x: x+5
display_side_by_side(df, df.apply(f), df.apply(f,axis=1))
```
