```python
import pandas as pd
import matplotlib.pyplot as plt

# CSV 파일 불러오기
df = pd.read_csv("서울시_민간소비_소득_통합.csv", encoding='utf-8-sig')

# 그래프 크기 설정
plt.figure(figsize=(14, 6))

# 막대 너비와 x축 인덱스
bar_width = 0.4
x = df['연도']
x_index = range(len(x))

# 막대 그래프 그리기
plt.bar(
    [i - bar_width/2 for i in x_index],
    df['민간소비지출액(천원)'],
    width=bar_width,
    label='민간소비지출액(천원)',
    color='royalblue'
)

plt.bar(
    [i + bar_width/2 for i in x_index],
    df['1인당_개인소득(천원)'],
    width=bar_width,
    label='1인당 개인소득(천원)',
    color='orange'
)

# X축 레이블 설정
plt.xticks(ticks=x_index, labels=x, rotation=45)
plt.xlabel('연도')
plt.ylabel('금액 (천 원)')
plt.title('서울시 연도별 민간소비지출액 및 1인당 개인소득 비교')
plt.legend()
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# CSV 파일 불러오기 (코랩에서는 직접 업로드 필요)
df = pd.read_csv('/content/주요지표_연간지표__20250506182224.csv', encoding='cp949')

# 연도별 데이터를 긴 형태로 변환
df_melted = df.melt(id_vars='계정항목별', var_name='년도', value_name='값')

# 특정 항목 선택 (예: 국내총생산(명목, 원화표시))
target = '국내총생산(명목, 원화표시) (십억원)'
plot_df = df_melted[df_melted['계정항목별'] == target].copy()
plot_df['년도'] = plot_df['년도'].astype(int)

# 그래프 출력
plt.figure(figsize=(12, 6))
plt.plot(plot_df['년도'], plot_df['값'], marker='o', linewidth=2)
plt.title(f'{target} 변화 추이', fontsize=14)
plt.xlabel('년도')
plt.ylabel('값 (십억원)')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# CSV 파일 불러오기
df = pd.read_csv('/content/주요지표_연간지표__20250506182224.csv', encoding='cp949')

# 데이터 정리 (긴 형식으로 변환)
df_melted = df.melt(id_vars='계정항목별', var_name='년도', value_name='값')
df_melted['년도'] = df_melted['년도'].astype(int)

# 각각 따로 그래프 그릴 항목
targets = {
    '국내총생산(명목, 원화표시) (십억원)': '국내총생산 ',
    '국민총소득(명목, 원화표시) (십억원)': '국민총소득 '
}

# 각각 그래프 생성
for indicator, title in targets.items():
    data = df_melted[df_melted['계정항목별'] == indicator]
    plt.figure(figsize=(12, 6))
    plt.plot(data['년도'], data['값'], marker='o', color='steelblue')
    plt.title(title, fontsize=14)
    plt.xlabel('년도')
    plt.ylabel('값 (십억원)', rotation=0, labelpad=50)  # y축 제목 가로로 설정
    plt.grid(True)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```
